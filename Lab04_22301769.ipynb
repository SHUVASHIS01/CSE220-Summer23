```python
class Patient:
  def __init__(self, id, name, age, bg):
    self.id = id
    self.name = name
    self.age=age
    self.bg = bg

class WRM:
  def __init__(self):
    self.head= None

  def register_patient(self, id, name, age, bg):
    new_p=Patient(id, name, age, bg)
    if self.head is None:
      self.head= new_p
      new_p.next = new_p
      new_p.prev = new_p
    else:
      last_p = self.head.prev
      last_p.next = new_p
      new_p.prev = last_p
      new_p.next = self.head
      self.head.prev = new_p

  def serve_patient(self):
    if self.head is None:
      print("No patients in the waiting room.")
      return
    serving_p = self.head
    if self.head.next == self.head:
      self.head= None
    else:
      self.head= self.head.next
      self.head.prev = serving_p.prev
      serving_p.prev.next= self.head
    print("Serving patient:", serving_p.name)

  def cancel_all(self):
    if self.head is None:
      print("No patients in the waiting room.")
      return
    self.head = None
    print("All appointments canceled. Doctor can go home")

  def can_doctor_go_home (self):
    if self.head is None:
      return True
    return False

  def show_all_patients(self):
    if self.head is None:
      print("No patients in the waiting room.")
      return
    current_p = self.head
    print(" Waiting patients: ")
    while True:
      print("ID:", current_p.id, "Name:", current_p.name)
      current_p = current_p.next
      if current_p == self.head:
        break

def display_menu():
  print("1. Add Patient")
  print("2. Serve Patient")
  print("3. Show All Patients")
  print("4. Can Doctor Go Home?")
  print("5. Cancel All Appointments")
  print("6. Exit")
wrm = WRM()
while True:
  display_menu()
  choice = input("Enter your choice: ")
  if choice == "1":
    id = int(input("Enter patient ID: "))
    name = input("Enter patient name: ")
    age = int(input("Enter patient age: "))
    bg = input("Enter patient blood group")
    wrm.register_patient (id, name, age, bg)
    print("Patient added successfully.")
  elif choice == "2":
    wrm.serve_patient()
  elif choice == "3":
    wrm.show_all_patients()
  elif choice == "4":
    if wrm.can_doctor_go_home():
      print("Yes, the doctor can go home.")
    else:
      print("No, the doctor cannot go home ")
  elif choice == "5":
    wrm.cancel_all()
  elif choice == "6":
    print("Exiting the program.")
    break
  else:
    print("Invalid choice. Please try again.")
```
