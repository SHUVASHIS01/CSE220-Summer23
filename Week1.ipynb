```python
# 3 dimentional array using python
array_3d=[None]*5
for i in range (len(array_3d)):
  array_3d[i]=[None]*5
  for j in range (len(array_3d[i])):
    array_3d[i][j]=[None]*5
print (array_3d)
```

```python
import numpy as np
arr = np.zeros((4,5,6,), dtype='int')
for i in range (arr.shape[0]):
  for j in range (arr.shape[1]):
    for k in range (arr.shape[2]):
      arr[i][j][k]=i*100+j*10+k
print (arr)
```

```python
arr1=np.array([[1,2,3],[4,5,6]])
arr2=np.array([[7,8],[9,10],[11,12]])
if arr1.shape[1]!=arr2.shape[0]:
  print('Not possible')
else:
  m=arr1.shape[0]
  n=arr1.shape[1]
  p=arr2.shape[1]
  result=np.zeros((m,p))
  for i in range (m):
    for j in range (p):
      s=0
      for k in range (n):
        s+=arr1[i][k]*arr2[k][j]
      result[i][j]=s
  print(result)
```
