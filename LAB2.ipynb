```python
import numpy as np
def removeElem(arr1,elem):
  c=0
  for i in arr1:
    if i==elem:
      c+=1
  arr2=np.array([None]*(len(arr1)-c))
  count=0
  for j in range (len(arr1)):
    if arr1[j]!=elem:
      arr2[count]=arr1[j]
      count+=1
    
  return arr2


arr1=np.array([ 3,5,8,9,3,6,3,3 ])
elem=int(input())
arr1=removeElem(arr1,elem)
print(arr1)
```
