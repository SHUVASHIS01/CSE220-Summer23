```python
def fact(n):
  if n==1:
    return 1
  else:
    return n*fact(n-1)
print(fact(int(input())))
```

```python
def fib(n):
  if n==1 or n==2:
    return 1
  else:
    return fib(n-1)+fib(n-2)
print(fib(7))
```

```python
def sumdigits(n):
  if n<10:
    return n
  else:
    return n%10 + sumdigits(n//10)
print(sumdigits(int(input())))
```

```python
def ears(n):
  if n==0:
    return 0
  elif n%2==1:
    return 2 + ears(n-1)
  elif n%2==0:
    return 3 + ears(n-1)
print(ears(int(input())))
```

```python
def count7(n):
  if n<10 and n!=7:
    return 0
  elif n%10==7:
    return 1+count7(n//10)
  else:
    return count7(n//10)
print(count7(7787))
```

```python
def countX(s):
    if len(s) == 0:
        return 0
    elif s[0] == 'x':
        return 1 + countX(s[1:])
    else:
        return countX(s[1:])

print(countX("xxhixx"))

```
