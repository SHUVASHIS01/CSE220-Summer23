```python
#You must run this cell to print matrix and for the driver code to work
def print_matrix(m):
  row,col = m.shape
  for i in range(row):
    c = 1
    print('|', end='')
    for j in range(col):
      c += 1
      if(len(str(m[i][j])) == 1):
        print(' ',m[i][j], end = '  |')
        c += 6
      else:
        print(' ',m[i][j], end = ' |')
        c += 6
    print()
    print('-'*(c-col))

```

Task 1: Zigzag Walk

```python
# Task 1
def walk_zigzag(floor):
  r,c= floor.shape
  for i in range (r):
    if i%2==0:
      for j in range (0,c,2):
        print(floor[i][j],end=' ')
      print()
    else:
      for k in range (c-1,-1,-2):
        print(floor[i][k],end=' ')
      print()
  pass



floor = np.array([[ '3W' , '8B' , '4W' , '6B' , '1W' , '5B'],['3B' , '2W' , '1B' , '6W' , '3B' , '8W'],
                  ['9W' , '0B' , '7W' , '5B' , '3W' , '8B'],['2B' , '1W' , '3B' , '6W' , '0B' , '4W'],
                   ['1W' , '4B' , '2W' , '8B' , '6W' , '6B']])
print_matrix(floor)
print('Walking Sequence:')
walk_zigzag(floor)
#This should print
# 3W 4W 1W
# 8W 6W 2W
# 9W 7W 3W
# 4W 6W 1W
# 1W 2W 6W
```

Task 2: Landscape Screen

```python
# Task 02
def landscape(m):
  r,c= m.shape
  arr=np.zeros((c,r),dtype='i')
  for i in range (r):
    for j in range (c):
      arr[j][i]=m[i][j]
  return arr


m = np.random.randint(15,size=(4,3)) #This creates an np array of 4 x 3 where the values are random numbers from [0,15) range
print_matrix(m)
print()
print_matrix(landscape(m))
```

Task 3: Seating Arrangement

```python
def arrangement(m, diff):
  r,c=m.shape
  count=0
  for i in range (r):
    for j in range (c):
      if i!=j:
        if abs(m[i][j]-m[j][i])==diff:
          pass
        else:
          count+=1
  if count>0:
    return False
  else:
    return True


m = np.array([[7,13,9,14],[12,8,15,11],[10,17,3,2],[15,10,1,4]])
print_matrix(m)
print()
returned_value = arrangement(m,1)
print(returned_value)#This should print False
unittest.output_test(returned_value, False)
```

Task 4: Chess Piece

```python
def show_knight_move(knight):
  arr=np.zeros((8,8),dtype='i')
  r,c=arr.shape
  kr,kc=knight
  arr[kr][kc]=33
  for i in range (r):
    for j in range (c):
      if (i==(kr-2) or i==(kr+2 ))and (j==(kc-1) or j==(kc+1)):
        arr[i][j]=3
      elif (i==(kr-1) or i==(kr+1))and (j==(kc-2) or j==(kc+2)):
        arr[i][j]=3

  return arr


knight = (3,4)
chess_board = show_knight_move(knight)
print_matrix(chess_board)
```

```python
def topple(knight,rook):
  arr=np.zeros((8,8),dtype='i')
  r,c=arr.shape
  kr,kc=knight
  rr,rc=rook
  arr[kr][kc]=33
  arr[rr][rc]=5
  rcount=0
  kcount=0
  count=0
  v=''
  for i in range (r):
    for j in range (c):
      if (i==(kr-2) or i==(kr+2))and (j==(kc-1) or j==(kc+1)):
        arr[i][j]=3
        if (i==rr) and j==rc :
          kcount+=1
        else:
          count+=1
      elif (i==(kr-1) or i==(kr+1))and (j==(kc-2) or j==(kc+2)):
        arr[i][j]=3
        if (i==rr) and j==rc :
          kcount+=1
        else:
          count+=1
      elif (i==rr and i==kr) or (j==rc and j==kc):
        rcount+=1
  if rcount>0:
    v='Rook can kill'
  elif count>0:
    v= 'cannot kill'
  elif kcount>0:
    v='knight can kill'


  return (arr, v )


knight = (5,6)
rook = (5,1)
chess_board, ans = topple(knight,rook)
print_matrix(chess_board)
print(ans) #This should print Rook can kill
```
