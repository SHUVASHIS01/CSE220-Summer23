```python
#You must run this cell to print matrix and for the driver code to work
def print_matrix(m):
  row,col = m.shape
  for i in range(row):
    c = 1
    print('|', end='')
    for j in range(col):
      c += 1
      if(len(str(m[i][j])) == 1):
        print(' ',m[i][j], end = '  |')
        c += 6
      else:
        print(' ',m[i][j], end = ' |')
        c += 6
    print()
    print('-'*(c-col))
```

# **Flip Vertically**

```python
import numpy as np

def flip_vertical(matrix):
  r,c=matrix.shape
  arr=np.zeros((r,c), dtype='int')
  x=r-1
  for i in range (r):
    for j in range (c):
      arr[i][j]=matrix[x][j]
    x-=1

  return arr

#Driver Code
#DO NOT CHANGE ANYTHING BELOW
print("=======TEST 1=======")
m = np.random.randint(15,size=(4,4)) #This creates an np array of 4 x 3 where the values are random numbers from [0,15) range
print_matrix(m)
print()
print_matrix(flip_vertical(m))
print()

print("=========TEST 2=========")
m = np.random.randint(15,size=(5,5)) #This creates an np array of 5 x 4 where the values are random numbers from [0,15) range
print_matrix(m)
print()
print_matrix(flip_vertical(m))
```
